```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Office_data = pd.read_csv("datasets/office_episodes.csv")
```

```python
Office_data
```

```python
plt.scatter(Office_data['episode_number'], Office_data['viewership_mil'],)
```

```python
# function to map rating to color
def rat2col(rating):
    if rating < 0.25:
        return("red")
    elif 0.25 <= rating < 0.50:
        return("orange")
    elif 0.50 <= rating < 0.75:
        return("lightgreen")
    else:
        return("darkgreen")
    
# Iterate over scaled_ratings row of Office_data to input color name to the colors list
colors = Office_data['scaled_ratings'].apply(rat2col)

# Inspect colors  
colors
```

```python
plt.scatter(Office_data['episode_number'], 
            Office_data['viewership_mil'],
            color = colors
           )
```

```python
# Iterate over has_guests row of Office_data to input size to the sizes list
sizes = Office_data['has_guests'].apply(lambda x: 250 if x else 25 )

# Inspect colors  
sizes
```

```python
plt.scatter(Office_data['episode_number'], 
            Office_data['viewership_mil'],
            color = colors,
            sizes = sizes
           )
```

### let's make the final plot bigger and label things

```python
# making the figure bigger
fig = plt.figure()
plt.rcParams['figure.figsize'] = [11, 7]
plt.scatter(Office_data['episode_number'], 
            Office_data['viewership_mil'],
            color = colors,
            sizes = sizes
           )

# setting the title 
plt.title("Popularity, Quality, and Guest Appearances on the Office")

# setting the labels
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")

plt.show()
```
